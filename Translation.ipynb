```python
import matplotlib.pyplot as plt
def rotate(x1,x2,y1,y2,tx,ty):
    x_list=[x1,x2]
    y_list=[y1,y2]
    plt.plot(x_list,y_list,marker="o",label="Before translation")
    x_list=[x1+tx,x2+tx]
    y_list=[y1+ty,y2+ty]
    plt.plot(x_list,y_list,marker="o",label="After translation")
    plt.title("Translation")
    plt.legend()
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    plt.show()
print("Enter line co-ordinates:")
x1,y1=map(int,input("Enter the point x1 and y1 :").split())
x2,y2=map(int,input("Enter the point x2 and y2 :").split())
tx,ty=map(int,input("Enter the tranlation factor tx,ty :").split())

rotate(x1,x2,y1,y2,tx,ty)
```
