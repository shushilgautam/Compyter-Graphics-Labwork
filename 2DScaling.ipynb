```python
import matplotlib.pyplot as plt
import os
print(os.getcwd())
def Scaling(polygon,sx,sy):
    x, y = zip(*polygon)
    plt.plot(x + (x[0],), y + (y[0],), color='black',linewidth=2)
    plt.fill(x,y,color='blue',label="Before Scaling")
    print("Before Scaling", end="")
    print(polygon)
    for i in range(len(polygon)):
        polygon[i][0]*=sx
        polygon[i][1]*=sy
    x, y = zip(*polygon)
    plt.plot(x + (x[0],), y + (y[0],), color='black',linewidth=2)
    plt.fill(x,y,color='green',label="After Scaling")
    print("After Scaling", end="")
    print(polygon)
    
# x1,y1=map(int,input("Enter the point x1 and y1 :").split())
# x2,y2=map(int,input("Enter the point x2 and y2 :").split())
polygon=[[2,4],[4,3],[3,6]]
sx,sy=map(int,input("Enter Scaling factors sx and sy : ").split())
Scaling(polygon,sx,sy)
plt.title("Scaling")
plt.legend()
plt.xlabel("x-axis")    
plt.ylabel("y-axis")
plt.show()
```
