```python
import os
print(os.getcwd())
import matplotlib.pyplot as plt
from copy import deepcopy
def shearing_x(polygon,shx,ax2):
    polygon[2][0]+=shx
    polygon[3][0]+=shx
    x, y = zip(*polygon)
    ax2.plot(x + (x[0],), y + (y[0],),color='black')
    ax2.fill(x,y,color='green',label="Xshear rectangle")
    
def shearing_y(object,shy,ax3):
    object[1][1]+=shy
    object[2][1]+=shy
    x, y = zip(*object)
    ax3.plot(x + (x[0],), y + (y[0],),color='black')
    ax3.fill(x,y,color='red',label="Yshear rectangle")
    
rectangle=[[0,0],[2,0],[2,4],[0,4]]
x, y = zip(*rectangle)
fig, (ax1, ax2,ax3) = plt.subplots(3, 1,figsize=(4, 6)) 
ax1.plot(x + (x[0],), y + (y[0],),color='black')
ax1.fill(x,y,color='blue',label="Original rectangle")
ax1.set_title("Shearing")
shx=int(input("Enter shear factor shx along x-axis"))
shy=int(input("Enter shear factor shy along y-axis"))
copy1=deepcopy(rectangle)
copy2=deepcopy(rectangle)
print(rectangle)
shearing_x(copy1,shx,ax2)
shearing_y(rectangle,shy,ax3)

ax1.legend()
ax2.legend()
ax3.legend()
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1,
                   wspace=0.3, hspace=0.3)
# plt.grid(True)
plt.show()
```
