```python
import os
print(os.getcwd())
import matplotlib.pyplot as plt
x1,y1=map(float,input("Enter the co-ordinate x1 y1 = ").split())
x2,y2=map(float,input("Enter the co-ordinate x2 y2 = ").split())
dx=abs(x2-x1)
dy=abs(y2-y1)
# slope m=
m=(y2-y1)/(x2-x1)
x=x1
y=y1
#initial decision parameter = p
i=1
x_list=[x]
y_list=[y]
print("k\tx\ty\tpk\t(x+1,y+1)")
if m<=1:
    p=2*dy-dx
    while(x<x2 and y<y2):
        print(i,"\t",x,"\t",y,"\t", p, "\t", end=" ")
        x+=1
        if p<0:
            p=p+2*dy
        else:
            y+=1
            p=p+(2*dy)-(2*dx)
        print(f"{(x,y)}")
        i+=1
        x_list.append(x)
        y_list.append(y)
        
else:
    p=2*dx-dy
    while(x<x2 and y<y2):
        print(i,"\t",x,"\t",y,"\t", p, "\t", end=" ")
        y+=1
        if p<0:
            p=p+2*dx
        else:
            x+=1
            p=p+(2*dx)-(2*dy)
        print(f"{(x,y)}")
        i+=1
        x_list.append(x)
        y_list.append(y)
plt.plot(x_list,y_list,marker="o")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Bresenham Line Drawing Algorithm")
plt.show()
print("Sushil Gautam")
```
