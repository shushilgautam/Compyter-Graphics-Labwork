```python
import matplotlib.pyplot as plt
from math import sin,cos,radians
def rotateFromOrigin(x1,y1,x2,y2,angle):
    x_list=[x1,x2]
    y_list=[y1,y2]
    plt.plot(x_list,y_list,marker="o",label="Before rotation")
    angle=radians(angle)
    x_list[0]=round(x1*cos(angle)-y1*sin(angle))
    y_list[0]=round(x1*sin(angle)+y1*cos(angle))
    x_list[1]=round(x2*cos(angle)-y2*sin(angle))
    y_list[1]=round(x2*sin(angle)+y2*cos(angle))
    plt.plot(x_list,y_list,marker="o",label="Rotation about origin")
        
    
x1,y1=map(int,input("Enter the point x1 and y1 :").split())
x2,y2=map(int,input("Enter the point x2 and y2 :").split())
angle=int(input("Enter the angle in degree :"))
# ax,ay=map(int,input("Enter the arbitrary point ax and ay :").split())
rotateFromOrigin(x1,y1,x2,y2,angle)
plt.legend()
plt.show()
```
