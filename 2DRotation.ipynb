```python
import matplotlib.pyplot as plt
from math import sin,cos,radians
import os
print(os.getcwd())
def rotateFromOrigin(x1,y1,x2,y2,angle,ax,ay):
    x_list=[x1,x2]
    y_list=[y1,y2]
    print("Point Before Rotation :")
    print(f"{x_list[0],y_list[0]} {x_list[1],y_list[1]}")
    plt.plot(x_list,y_list,marker="o",label="Before rotation")
    # rotation about origin is given in formula
    angle=radians(angle)
    x_list[0]=round(x1*cos(angle)-y1*sin(angle))
    y_list[0]=round(x1*sin(angle)+y1*cos(angle))
    x_list[1]=round(x2*cos(angle)-y2*sin(angle))
    y_list[1]=round(x2*sin(angle)+y2*cos(angle))
    print("Point After Rotation about origin:")
    print(f"{x_list[0],y_list[0]} {x_list[1],y_list[1]}")
    plt.plot(x_list,y_list,marker="o",label="Rotation about origin")
    # rotation about arbiratory point = A'+ arbiratory point
    x_list[0]+=ax
    y_list[0]+=ay
    x_list[1]+=ax
    y_list[1]+=ay
    print("Point After Rotation about arbitrary point : ")
    print(f"{x_list[0],y_list[0]} {x_list[1],y_list[1]}")
    plt.plot(x_list,y_list,marker="o",label="Rotation about arbitrary point")    
    
x1,y1=map(int,input("Enter the point x1 and y1 :").split())
x2,y2=map(int,input("Enter the point x2 and y2 :").split())
angle=int(input("Enter the angle in degree :"))
ax,ay=map(int,input("Enter the arbitrary point ax and ay :").split())
rotateFromOrigin(x1,y1,x2,y2,angle,ax,ay)
plt.legend()
plt.title("Rotation")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
