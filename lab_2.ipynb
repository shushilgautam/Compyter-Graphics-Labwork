```python
import matplotlib.pyplot as plt
x1,y1=map(float,input("Enter co-ordinate x1 and y1 : ").split())
x2,y2=map(float,input("Enter co-ordinate x2 and y2 : ").split())
dx=x2-x1
dy=y2-y1
if(abs(dy)>abs(dx)):
    steps=dy
else:
    steps=dx
x_list=[]
y_list=[]
xincr = dx/float(steps)
yincr = dy/float(steps)
for i in range(1,int(steps)):
    x1=x1+xincr
    y1=y1+yincr
    x_list.append(x1)
    y_list.append(y1)
plt.plot(x_list,y_list,marker="o")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("DDA Algorithm")
plt.show()
print("Name : Sushil Gautam")
print("Roll no: 19")
    
```
