```python
import os
print(os.getcwd())
import matplotlib.pyplot as plt
x1,y1=map(float,input("Enter co-ordinate x1 and y1 : ").split())
x2,y2=map(float,input("Enter co-ordinate x2 and y2 : ").split())
dx=x2-x1
dy=y2-y1
if(abs(dy)>abs(dx)):
    steps=dy
else:
    steps=dx
x_list=[round(x1)]
y_list=[y1]
xincr = dx/float(steps)
yincr = dy/float(steps)
print("steps\txk\tyk\t(xk+1,yk+1)\tRound off")
for i in range(0,int(steps)):
    print(i,"\t","%.2f"%x1,"\t",y1,"\t",end="")
    x1=x1+xincr
    y1=y1+yincr
    x_list.append(round(x1))
    y_list.append(y1)
    print(f"{(round(x1,2),y1)}\t{round(x1),round(y1)}")
plt.plot(x_list,y_list,marker="o")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("DDA Algorithm")
plt.show()
print("Name : Sushil Gautam")
print("Roll no: 19")
```
