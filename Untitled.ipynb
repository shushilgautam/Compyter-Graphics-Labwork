```python
import matplotlib.pyplot as plt
x1,y1=map(float,input("Enter the co-ordinate x1 y1 = ").split())
x2,y2=map(float,input("Enter the co-ordinate x2 y2 = ").split())
dx=abs(x2-x1)
dy=abs(y2-y1)
#decision parameter = p
p=2*dy-dx
if(x2>x1):
    x=x1
    y=y1
    x_end=x2
    y_end=y2
else:
    x=x2
    y=y2
    x_end=x1
    y_end=y1
x_list=[x]
y_list=[y]
print("k\tx\ty\tp\t(x+1,y+1)")
i=0
while(x<x_end and y<y_end):
    print(i,"\t",x,"\t",y,"\t",end=" ")
    i=i+1
    x=x+1
    if(p<0):
        p=p+(2*dy)
    else:
        y=y+1
        p=p+(2*dy)-(2*dx)
    print(p,"\t",x," ",y)
    x_list.append(x)
    y_list.append(y)
    
print(x_list)
print(y_list)
plt.plot(x_list,y_list,marker="o")
plt.show()
```
