```python
import os
print(os.getcwd())
import matplotlib.pyplot as plt
def bezier_curve(x,y):
    xlist,ylist=[x[0]],[y[0]]
    t=0.0
    while t<1.0:
        t+=0.0005
        xt=(1-t)**3*x[0]+3*t*((1-t)**2)*x[1]+3*t*t*(1-t)*x[2]+t**3*x[3]
        yt=(1-t)**3*y[0]+3*t*((1-t)**2)*y[1]+3*t*t*(1-t)*y[2]+t**3*y[3]
        xlist.append(xt)
        ylist.append(yt)
    return xlist,ylist
# main program start
x,y=[],[]
print("Enter 4 Control point of line :")
# fig, [ax, ax1] = plt.subplots(2, 1)
for i in range(4):
    tempx,tempy=map(float,input(f"Enter point x{i} y{i} :").split())
    x.append(tempx)
    y.append(tempy)
plt.plot(x,y,marker="o",label="Original Line",linestyle="dashed")
xlist,ylist=bezier_curve(x,y)
plt.plot(xlist,ylist,label="Curved Line")
plt.suptitle("Cubic Bezier Curve")
plt.legend()
plt.grid(True)
plt.show()
```
